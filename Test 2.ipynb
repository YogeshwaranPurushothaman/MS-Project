```python
import pandas as pd

# Specify the file path
file_path = r"C:\Users\Bhuvanesh\Desktop\Yogesh Guvi\Microsoft -Classifying Cybersecurity Incidents\Data Set\GUIDE_Train.csv"

# Set the chunk size to load the file in manageable parts
chunksize = 10**6  # 1 million rows per chunk (adjust as necessary)

# Process each chunk
chunk_list = []  # To store processed chunks

for chunk in pd.read_csv(file_path, chunksize=chunksize, low_memory=False):
    chunk_list.append(chunk)


# Optionally concatenate all chunks back into a single DataFrame
df = pd.concat(chunk_list, ignore_index=True)

# Continue processing your data
```
